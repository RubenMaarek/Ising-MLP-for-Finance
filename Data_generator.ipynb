```python
import random

# Function to plot returns for a given temperature
def plot_returns(temperature, returns):
    plt.figure()
    plt.plot(returns)
    plt.title(f"Returns for Temperature: {temperature}")
    plt.xlabel("Epoch")
    plt.ylabel("Return")
    plt.show()

# Open the HDF5 file
filename = 'returns_dataset.h5'
with h5py.File(filename, 'r') as f:
    
    # Get all temperatures
    temperatures = list(f.keys())
    
    # Select 5 random temperatures
    random_temperatures = random.sample(temperatures, 5)
    
    # Plot returns for selected temperatures
    for temp in random_temperatures:
        # Get returns for the current temperature
        returns = f[temp][:]
        
        # Plot returns
        plot_returns(temp, returns)
```
